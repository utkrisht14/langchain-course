# Upsert text

In Pinecone, **“upsert”** is a combination of **“update” + “insert.”**

It means:

> 👉 *Add new vectors to your index, or update them if they already exist.*

---

## 🧩 How it works

When you call `upsert` in Pinecone:

* If the **ID does not exist** → it **inserts** a new vector
* If the **ID already exists** → it **updates/replaces** that vector

So you don’t need to check beforehand whether something exists — Pinecone handles that automatically.

---

## 🧠 Simple example

```python
index.upsert([
    ("vec1", [0.1, 0.2, 0.3]),
    ("vec2", [0.4, 0.5, 0.6])
])
```

* If `"vec1"` is new → it gets added
* If `"vec1"` already exists → its values get overwritten

---

## 📌 Why it’s useful

* Saves you from writing separate **insert vs update logic**
* Keeps your vector database **in sync with changing data**
* Common in pipelines where embeddings are re-generated

---

## ⚠️ One important note

Upsert **replaces the entire vector (and metadata)** for that ID.
It’s not a partial update — think of it as “overwrite if exists.”
