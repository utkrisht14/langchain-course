## Sematic Search

**Semantic search** is a search technique that focuses on the **meaning and intent behind a query**, rather than just matching exact keywords.

---

## 🧠 What makes it different

Traditional keyword search looks for exact word matches.
Semantic search, on the other hand, understands:

* context
* synonyms
* user intent

For example:

* Query: *“best place to buy affordable laptop”*
* Semantic search might return results about *cheap notebooks* or *budget computers* — even if those exact words weren’t used.

---

## ⚙️ How it works (high level)

Semantic search typically relies on **embeddings** — numerical vector representations of text.

* Text (query or document) → converted into vectors using models
* Similar meanings → vectors are closer in space
* Search → finds vectors that are closest to the query vector

This is why vector databases like Pinecone are commonly used — they efficiently store and search these embeddings.

---

## 📌 Key benefits

* More relevant results (based on meaning, not just words)
* Handles synonyms and paraphrases well
* Works better for natural language queries
* Essential for modern AI applications (chatbots, RAG systems)

---

## ⚠️ Limitations

* Requires embedding models (extra computation)
* May miss exact keyword matches if not tuned properly
* Needs good data representation for best results

---

## 🧩 Where it’s used

* AI search engines
* Question-answering systems
* Recommendation systems
* Chatbots and assistants

---

## ✅ In one line

> Semantic search finds results based on **what you mean**, not just **what you type**.


Search the dense index for ten records that are most semantically similar to the query, “Famous historical structures and monuments”.

Again, because your index is integrated with an embedding model, you provide the query as text and Pinecone converts the text to a dense vector automatically.


```python
from c_upsert_text import dense_index
```

```python
# Define the query
query = "Famous historical structures and monuments"

# Search the dense index
results = dense_index.search(
    namespace="example-namespace",
    query = {
        "top-k": 10,
        "inputs": {
            "text": query
        }
    }
)
```

```python
# Print the results
for hit in results["result"]["hits"]:
    print(f"id: {hit['_id']:<5} | score: {round(hit['_score'], 2):<5} | category: {hit['fields']['category']:<10} | text: {hit['fields']['chunk_text']:<50}")
```
